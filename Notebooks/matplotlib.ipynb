```python
data={'salary':[10000,20000,30000,40000,100000,2000,53200,45600]}
print(data)
newdata=pd.DataFrame(data)
newdata.shape
newdata.loc[8]=[100000]
print(newdata)
plt.plot(newdata , color='green', marker='o', linestyle=':')
plt.grid()
plt.show()
```

```python
plt.boxplot(newdata['salary'])
plt.show()
```

```python
# Add a new column named 'department' for practicing pie chart creation
newdata['department'] = ['Sales', 'Marketing', 'Engineering', 'Finance', 'Sales', 'Marketing', 'Engineering', 'Finance', 'Engineering']

# Display the updated DataFrame
display(newdata)
```

```python
count1=newdata['department'].value_counts()
count2=newdata['salary'].value_counts()
print(count1)
print('//'*20)
print(count2)
```

```python
plt.pie(count1, labels=count1.index, autopct='%1.3f%%', explode=(0,0,0,0.1))
plt.show()
```

```python
plt.bar(count1.index,count1, color=['green','blue','red'])
plt.show()
```

```python
# Add an 'age' column for scatterplot
newdata['age'] = [25, 30, 35, 40, 50, 22, 38, 42, 55]

# Display the updated DataFrame
display(newdata)
```

```python
plt.bar(newdata['age'],newdata['salary'],color='black')
plt.show()
```

```python
salary_dept=newdata.groupby('department')['salary'].sum()
salary_dept
```

```python
plt.pie(salary_dept,labels=salary_dept.index , autopct='%1.1f%%' , explode=[0.1,0,0,0] , shadow=True
 )
plt.show()
```

```python
eng_sal=newdata[newdata['department']=='Engineering']['salary']
fin_sal=newdata[newdata['department']=='Finance']['salary']
mar_sal=newdata[newdata['department']=='Marketing']['salary']
sal_sal=newdata[newdata['department']=='Sales']['salary']
```

```python
dept=newdata['department'].unique()
mean_sal=[sal_sal.mean(),mar_sal.mean(),eng_sal.mean(),fin_sal.mean()]
```

```python
plt.bar(dept,mean_sal ,  width=0.4 , color='#301934')
plt.xlabel("Department")
plt.ylabel("Mean Salary")
plt.title("Mean Salary BY Department")
plt.show()
```

```python
sort_age=newdata.sort_values(by='age')
sort_age
```

```python
fig ,axs=plt.subplots(1,3,figsize=(15,5))

#line plot

axs[0].plot(sort_age["age"],newdata['salary'])
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

#scatter plot

axs[2].scatter(sort_age["age"],newdata['salary'])
axs[2].grid()
axs[2].set_title("Scatter Plot")
axs[2].set_xlabel("Age")
axs[2].set_ylabel("Salary")

#bar plot

axs[1].bar(sort_age["age"],newdata['salary'])
axs[1].grid()
axs[1].set_title("Bar Plot")
axs[1].set_xlabel("Age")
axs[1].set_ylabel("Salary")

plt.tight_layout()
plt.show()
```
