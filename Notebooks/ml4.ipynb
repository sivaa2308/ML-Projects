```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
X=np.array([1,2,3,4,5,6])
Y=np.array([7,8,9,10,11,12])
```

```python
x_sm=sm.add_constant(X)
```

```python
model=sm.OLS(Y,x_sm).fit()
```

```python
Y_pred=model.predict(x_sm)
```

```python
model.summary()
```

```python
data=pd.read_csv('/content/2.01.+Admittance.csv')
```

```python
data
```

```python
data1=data
```

```python
data1['Admitted'] = data['Admitted'].str.strip().str.title().map({"No":0, "Yes":1})
```

```python
data1
```

```python
x1=data1[['SAT']]
y1=data1['Admitted']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

xtrain , xtest ,  ytrain , ytest=train_test_split(x1,y1,test_size=0.20, random_state=42)
logistic=LogisticRegression(max_iter=1000)

logistic.fit(xtrain,ytrain)
```

```python
y_pred=logistic.predict(xtest)
```
