```python
l1 = [1, 3, 5, 7, 9]
l2 = [2, 4, 6, 8, 10]
i, j = 0, 0

while i < len(l1) and j < len(l2):
    if l1[i] < l2[j]:
        print(l1[i])
        i += 1
    else:
        print(l2[j])
        j += 1

while i < len(l1):
    print(l1[i])
    i += 1

while j < len(l2):
    print(l2[j])
    j += 1
```

**Broadcasting**

Broadcasting is a NumPy feature that allows operations between arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one — without copying data unnecessarily.

```python
image=np.array([[200,250],[300,350]])
print(image.shape)
image+=50
print(image)
```

**np.nan**

```python
arr = np.array([1, 2, np.nan, 4])
b=[1,2,3,4,np.inf,5]
print(arr)
print(np.isnan(arr))
print(np.isinf(b))
arr=np.nan_to_num(arr,nan=0)
b=np.nan_to_num(b,posinf=10000,neginf=-100000)  # remove not a number or infinet values
print(arr)
print(b)
```

**np.power**

np.power(arr1, arr2) raises each element of arr1 to the power of the corresponding element in arr2, element-wise.

```python
import numpy as np

arr1 = np.array([2, 3, 4])
arr2 = np.array([3, 2, 1])

result = np.power(arr1, arr2)
print(result)
```
