```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tit= sns.load_dataset('titanic')
```

```python
tit
```

```python
tit.isna().sum()
```

```python
# If you want your model to know which rows had missing age before
tit['age_missing'] = tit['age'].isnull().astype(int)   # do this BEFORE filling
```

```python
tit.info()
```

```python
tit.shape
```

```python
tit[tit.isna().any(axis=1)]
```

```python
tit.drop('deck',axis=1,inplace=True)
```

```python
print(tit.groupby(['sex','class']))
```

```python
tit['age'] = tit.groupby(['sex','class'])['age'].transform(lambda x: x.fillna(x.median()))
tit=tit.dropna(axis=0)
```

```python
tit.shape
```

```python
tit
```

```python
tit=pd.get_dummies(tit,columns=['sex','class','embarked','who'],drop_first=True)
tit
```

```python
tit.columns.values
```

```python
tit.drop(['embark_town','alive'],inplace=True,axis=1)
```

```python
tit
```

```python
tit.columns.values
```

```python
X=tit.drop('survived',axis=1)
X=X.astype(int)
Y=tit['survived']
```

```python
print(X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
tit
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(X_test)
accuracy=accuracy_score(Y_test,y_pred)
print(f'accuracy: {accuracy}')
```
