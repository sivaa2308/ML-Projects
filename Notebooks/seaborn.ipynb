```python
penguins=sns.load_dataset("penguins")
print(penguins.head())
```

```python
penguins['species'].value_counts()
```

```python
sns.set_style('darkgrid')
sns.set_context("notebook", font_scale=1)
sns.scatterplot(data=penguins , x='bill_length_mm' , y='body_mass_g' , hue='species' , palette="Set1")
plt.title("Penguins: Bill Length vs Body Mass")
plt.show()
```

```python
sns.stripplot(x='species',y='body_mass_g' , data=penguins , hue='island' , palette="Set1" )
```

```python
sns.stripplot(x='species',y='body_mass_g' , data=penguins , hue='island' , palette="Set1" , dodge=True)
```

```python
sns.swarmplot(x='species',y='body_mass_g' , data=penguins , hue='island' , palette="Set1" )
```

```python
sns.swarmplot(x='species',y='body_mass_g' , data=penguins , hue='island' , palette="Set1" )
sns.despine()
```

```python
sns.histplot(data=penguins , x='body_mass_g' , hue='species' , palette='Set1' , multiple='stack' , kde=True)
```
