```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

train_url = "http://d2l-data.s3-accelerate.amazonaws.com/kaggle_house_pred_train.csv"
test_url = "http://d2l-data.s3-accelerate.amazonaws.com/kaggle_house_pred_test.csv"

df_train = pd.read_csv(train_url)
df_test = pd.read_csv(test_url)

print(df_train.shape)  # Expect (1460, 81)
print(df_train.head())
```

```python
missing = df_train.isna().sum()
missing = missing[missing != 0].sort_values(ascending=False)
print(missing)
```

```python
X=df_train.drop('SalePrice', axis=1)
Y=df_train['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_valid , y_train , y_valid = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
numeric_cols=x_train.select_dtypes(include=np.number).columns.tolist()
categorical_cols=x_train.select_dtypes(exclude=np.number).columns.tolist()
len(numeric_cols)
```

```python
len(categorical_cols)
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

x_train_num = num_imputer.fit_transform(x_train[numeric_cols])
x_valid_num = num_imputer.transform(x_valid[numeric_cols])

x_train_cat = cat_imputer.fit_transform(x_train[categorical_cols])
x_valid_cat = cat_imputer.transform(x_valid[categorical_cols])
```

```python
x_train_num
```

```python
x_train_num_df = pd.DataFrame(x_train_num , columns=numeric_cols , index= x_train.index)
x_valid_num_df = pd.DataFrame(x_valid_num , columns=numeric_cols , index= x_valid.index)

x_train_cat_df = pd.DataFrame(x_train_cat , columns=categorical_cols , index=x_train.index)
x_valid_cat_df = pd.DataFrame(x_valid_cat , columns=categorical_cols , index=x_valid.index)
```

**Hot Encoding**

```python
x_train_encoded = pd.get_dummies(x_train_cat_df)
x_valid_encoded = pd.get_dummies(x_valid_cat_df)
```

```python
x_train_encoded, x_valid_encoded = x_train_encoded.align(
    x_valid_encoded,
    join='left',
    axis=1,
    fill_value=0
)
```

```python
x_train_final = pd.concat([x_train_num_df , x_train_encoded] , axis=1)
x_valid_final = pd.concat([x_valid_num_df , x_valid_encoded] , axis=1)
```

```python
x_train_final.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

model = RandomForestRegressor(n_estimators=100 , random_state=42)
model.fit(x_train_final , y_train)

y_pred = model.predict(x_valid_final)
mse = mean_squared_error(y_valid , y_pred)
print(mse)
```
