```python
import pandas as pd
```

```python
customers=pd.read_csv("/content/drive/MyDrive/ML_Resources/customers.csv")
orders=pd.read_csv("/content/drive/MyDrive/ML_Resources/orders.csv")
products=pd.read_csv("/content/drive/MyDrive/ML_Resources/products.csv")
ratings=pd.read_csv("/content/drive/MyDrive/ML_Resources/ratings.csv")
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
orders
```

```python
merged_data = orders.merge(products, on='product_id').merge(customers, on='customer_id')
```

```python
merged_data
```

```python
merged_data[merged_data['product_id']==38]
```

```python
merged_data['revenue']=merged_data['price']*merged_data['quantity']
merged_data
```

```python
product_perfomance=merged_data.groupby('product_name').agg({'revenue':'sum','quantity':'sum'}).reset_index()
product_perfomance
```

```python
top_product_revenue=product_perfomance.sort_values(by='revenue',ascending=False)
top_product_revenue
```

```python
top_product_quantity=product_perfomance.sort_values(by='quantity',ascending=False)
top_product_quantity
```

```python
latest_month=merged_data['order_date'].dt.to_period('M').max()
last_month=merged_data[merged_data['order_date'].dt.to_period('M')==latest_month]

top_clients_last_month = last_month.groupby('name').agg({'revenue':'sum'})
top_clients_last_month = top_clients_last_month.sort_values(by='revenue', ascending=False).reset_index()
top_clients_last_month
```

```python
import datetime as dt

latest_date = merged_data['order_date'].max() + dt.timedelta(days=1)

rfm=merged_data.groupby('customer_id').agg({'order_date':lambda x:(latest_date- x.max()).days ,
                                            'order_id':'count' ,
                                            'revenue':'sum'}).rename(columns={
    'order_date':'recency',
    'order_id':'frequency',
    'revenue':'monetary'
})
```

```python
print(rfm.columns)
```

```python
rfm['recency_score'] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['frequency_score'] = pd.qcut(rfm['frequency'], 5, labels=[1, 2, 3, 4, 5])
rfm['monetary_score'] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm['rfm_score'] = rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str) + rfm['monetary_score'].astype(str)

rfm_sorted = rfm.sort_values(by=['rfm_score'], ascending=False)
```
