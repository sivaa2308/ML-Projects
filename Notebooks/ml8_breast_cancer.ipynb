```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data , columns = data.feature_names)
df['target']=data.target
df.head()
```

```python
df.isna().sum()
```

```python
df.target.value_counts()
```

```python
x = df.drop('target' , axis = 1)
y = df.target
```

```python
from sklearn.model_selection import train_test_split

xtrain , xtest , ytrain , ytest = train_test_split(x , y , test_size=0.2 , random_state=42 , stratify=y)
```

```python
print(f"xtrain shape : {xtrain.shape}")
print(f"xtest shape : {xtest.shape}")
print(f"ytrain shape : {ytrain.shape}")
print(f"ytest shape : {ytest.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(xtrain)
x_test = scaler.transform(xtest)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

lr = LogisticRegression(max_iter = 1000)
dt = DecisionTreeClassifier(random_state = 42)
rf = RandomForestClassifier(random_state = 42)

lr.fit(x_train , ytrain)
dt.fit(x_train , ytrain)
rf.fit(x_train , ytrain)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

models={ 'LogisticRegression': lr , "DecisionTreeClassifier" : dt , "RandomForestClassifier" : rf}

for name , model in models.items():
  ypred = model.predict(x_test if name == "LogisticRegression" else xtest)

  print(f"{name} train Accuracy: {accuracy_score(ytest , ypred)}")
  print(f"{name} train confussion matrix \n{confusion_matrix(ytest , ypred)}")
  print(f"{name} train classification report {classification_report(ytest , ypred)}")
  print("*"*50)
```

```python
best_acc = 0
best_c = None

for c in [0.01, 0.1, 1, 10, 100]:
    clf = LogisticRegression(C=c, max_iter=5000)
    clf.fit(x_train, ytrain)
    acc = accuracy_score(ytest, clf.predict(x_test))
    print(f"C={c}, Accuracy={acc:.4f}")

    if acc > best_acc:
        best_acc = acc
        best_c = c

print("Best C:", best_c, "with Accuracy:", best_acc)
```

**Pipelines**

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

path = [("Scaler" , StandardScaler()) , ("logicreg",LogisticRegression(max_iter = 5000))]
pipe = Pipeline(path)

pipe.fit(xtrain , ytrain)

y_predict = pipe.predict(xtest)

accuracy = accuracy_score(ytest , y_predict)
print(f"Accuracy of LogisticRegression: {accuracy}")
print(f"Confusion Matrix : \n{confusion_matrix(ytest , y_predict)}")
print(f"Classification Report : \n{classification_report(ytest , y_predict)}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report

path2 = [("dectree",DecisionTreeClassifier(random_state = 42))]
pipe2 = Pipeline(path2)

pipe2.fit(xtrain , ytrain)
y_pred2 = pipe2.predict(xtest)

accuracy = accuracy_score(ytest , y_pred2)
print(f"Accuracy of DecisionTreeClassifier: {accuracy}")
print(f"Confusion Matrix : \n{confusion_matrix(ytest , y_pred2)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

pipe_rf = Pipeline([
    ("clf", RandomForestClassifier(random_state=42))
])

pipe_rf.fit(xtrain, ytrain)
y_pred_rf = pipe_rf.predict(xtest)

print("Random Forest Pipeline")
print("Accuracy of RandomForestClassifier:", accuracy_score(ytest, y_pred_rf))
print(confusion_matrix(ytest, y_pred_rf))
print(classification_report(ytest, y_pred_rf))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Pipeline
path = [("scaler", StandardScaler()), ("log", LogisticRegression(max_iter=5000))]
pipeline = Pipeline(path)

# Multiple param grids to avoid invalid solver-penalty combos
param = [
    {"log__C": [0.01, 0.1, 1, 10],
     "log__penalty": ["l2"],
     "log__solver": ["lbfgs"],
     "log__class_weight": [None, "balanced"]},

    {"log__C": [0.01, 0.1, 1, 10],
     "log__penalty": ["l1", "l2"],
     "log__solver": ["liblinear"],
     "log__class_weight": [None, "balanced"]},

    {"log__C": [0.01, 0.1, 1, 10],
     "log__penalty": ["l1", "l2", "elasticnet"],
     "log__solver": ["saga"],
     "log__class_weight": [None, "balanced"]},
]

# GridSearchCV
grid = GridSearchCV(pipeline, param, cv=5, scoring="accuracy", n_jobs=-1)
grid.fit(xtrain, ytrain)

# Predictions
ypred = grid.predict(xtest)

# Results
print("Best Params:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
print("Test Accuracy:", accuracy_score(ytest, ypred))
print("Confusion Matrix:\n", confusion_matrix(ytest, ypred))
print("Classification Report:\n", classification_report(ytest, ypred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Pipeline
dt_pipe = Pipeline([
    ("clf", DecisionTreeClassifier(random_state=42))
])

# Parameter grid for Decision Tree
dt_params = {
    "clf__criterion": ["gini", "entropy", "log_loss"],  # splitting rules
    "clf__max_depth": [3, 5, 10, None],                # tree depth
    "clf__min_samples_split": [2, 5, 10],              # min samples to split a node
    "clf__min_samples_leaf": [1, 2, 4],                # min samples per leaf
    "clf__class_weight": [None, "balanced"]            # handle imbalance
}

# GridSearchCV
grid_dt = GridSearchCV(dt_pipe, dt_params, cv=5, scoring="accuracy", n_jobs=-1)
grid_dt.fit(xtrain, ytrain)

# Predictions
y_pred_dt = grid_dt.predict(xtest)

# Results
print("\n=== Decision Tree with GridSearch ===")
print("Best Params:", grid_dt.best_params_)
print("Best CV Score:", grid_dt.best_score_)
print("Test Accuracy:", accuracy_score(ytest, y_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(ytest, y_pred_dt))
print("Classification Report:\n", classification_report(ytest, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
path = [("scaler",StandardScaler()),("rf",RandomForestClassifier(random_state=42))]
pipeline = Pipeline(path)

rf_param_grid = {
    "rf__n_estimators": [100, 200, 300],
    "rf__max_depth": [None, 5, 10, 20],
    "rf__min_samples_split": [2, 5, 10],
    "rf__min_samples_leaf": [1, 2, 4]
}

grid_rf = GridSearchCV(pipeline, param_grid=rf_param_grid , cv=5 , scoring="accuracy" , n_jobs=-1 , verbose=2)
grid_rf.fit(xtrain , ytrain)
y_pred_rf = grid_rf.predict(xtest)

print("Best Parameters:", grid_rf.best_params_)
print("Best CV Score:", grid_rf.best_score_)
print("Test Accuracy:", grid_rf.score(xtest, ytest))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# 1. Pipeline
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(probability=True, random_state=42))
])

# 2. Parameter grid
svm_param_grid = {
    "svm__C": [0.01, 0.1, 1, 10],
    "svm__kernel": ["linear", "rbf"],
    "svm__gamma": ["scale", "auto"]
}

# 3. RandomizedSearchCV
svm_grid = RandomizedSearchCV(
    pipeline,
    param_distributions=svm_param_grid,
    n_iter=5,
    scoring="accuracy",
    n_jobs=-1,
    random_state=42,
    verbose=2,
    cv=5
)

# 4. Fit model
svm_grid.fit(xtrain, ytrain)

# 5. Predict
ypredict = svm_grid.predict(xtest)

# 6. Results
print("Best Parameters:", svm_grid.best_params_)
print("Best CV Score:", svm_grid.best_score_)
print("Test Accuracy:", svm_grid.score(xtest, ytest))
```
