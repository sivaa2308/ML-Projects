```python
new_fruit_data = {
    "Fruit": [
        "Apple", "Banana", "Orange", "Kiwi", "Mango",
        "Papaya", "Grapes", "Pineapple", "Guava", "Blueberry"
    ],
    "Protein": [
        0.3, 1.1, np.nan, 2.1, np.nan,
        0.5, 0.6, np.nan, 0.9, 0.7
    ],
    "Calories": [
        52, 89, 62, np.nan, 60,
        np.nan, 67, 50, np.nan, 57
    ]
}

fn=pd.DataFrame(new_fruit_data)
fn
```

```python
fn.info()
```

```python
fn.describe()
```

```python
fn.iloc[3:6]
```

```python
print(fn.isnull().sum()) # This is used ti identufy the number of null values in each column
print("/"*30)
print("/"*30)
print(fn.notnull())
```

```python
fn.dropna(how='any') #in this droping , if nay value is nan it will remove that value and the column
```

```python
fn.dropna(how='all')  #In this code it will work only if , all the columns are nan , otherwise no removal
```

```python
fn.fillna(0)  #this is a methode for filling data in the nan
```

```python
print(fn['Protein'].mean())                #This is the methode for fillling the  nan using the mean of theat column
fn['Protein'].fillna(fn['Protein'].mean())
```

```python
print(fn['Calories'].mean())
fn['Calories'].fillna(fn['Calories'].mean())#This is the methode for fillling the  nan using the mean of theat column
```

```python
fn['Calories'].fillna(method='ffill') #This filling methode is forward fill , which fill the nan with the higher value
```

```python
fn['Calories'].fillna(method='bfill') #this is backward fill
```

```python
fn
```

```python
fn['Fruit']=fn['Fruit'].replace('Apple','Mango')
fn
```

```python
dupli=fn[fn.duplicated(keep='first ')]
dupli
```

```python
fn=fn.drop_duplicates(keep='first') # or keep='last' , for removing the duplicated elements
fn
```

```python

fn['Calories']=fn['Calories'].apply(lambda x: x*10 if x<10 else x)
fn['Calories']=fn['Calories'].fillna(fn['Calories'].mean())
fn['Calories']=fn['Calories'].round(3)
fn['Protein'] = fn['Protein'].fillna(fn['Protein'].mean())
fn['Protein'] = fn['Protein'].round(3)
fn
```
