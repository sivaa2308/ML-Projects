```python
import pandas as pd

train_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(train_url)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
df.drop('Cabin', axis = 1 , inplace = True)
df.shape
```

```python
df['title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df.title.unique()
```

```python
df['title']=df.title.replace({'Ms':'Miss', 'Mme':'Mrs' , 'Mlle':'Miss'})
```

```python
df.title.unique()
```

```python
df['title'] = df['title'].where(df['title'].isin(['Mr', 'Mrs', 'Miss', 'Ms']), 'Rare')
```

```python
df.title.unique()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Age'].describe()
```

```python
df['Age']=df.groupby('title')['Age'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0] , inplace=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
df.drop('Name', axis = 1 , inplace = True)
df.drop('Ticket', axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df , columns=['Sex' , 'Embarked' , 'title'] , drop_first=True)
```

```python
df.head()
```

```python
x=df.drop('Survived' , axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split

xtrain , xtest , ytrain, ytest = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
xtrain.shape , xtest.shape , ytrain.shape , ytest.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model=RandomForestClassifier(n_estimators=100 , random_state=42 ,max_depth=None )

model.fit(xtrain, ytrain)

ypred = model.predict(xtest)

accuracy = accuracy_score(ytest , ypred)
```

```python
print("Accuracy:", accuracy_score(ytest, ypred))
print("Classification Report:\n", classification_report(ytest, ypred))
```
