**Categorical data**

```python
df["Item"] = df["Item"].str.strip().str.lower().replace({"unknown": "Unknown", "error": "Unknown"})
df["Item"] = df["Item"].fillna("Unknown")
df["Payment Method"] = df["Payment Method"].fillna("Unknown")
df["Location"] = df["Location"].fillna("Unknown")
print(df["Item"].isna().sum())
print(df["Payment Method"].isna().sum())
print(df["Location"].isna().sum())
```

**Numerical data**

```python
df["Quantity"] = pd.to_numeric(df["Quantity"] , errors="coerce")
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"] , errors="coerce")
df["Total Spent"] = pd.to_numeric(df["Total Spent"] , errors="coerce")
```

```python
df["Quantity"] = df.groupby("Item")["Quantity"].transform(lambda x: x.fillna(x.mean()))
df["Price Per Unit"] = df.groupby("Item")["Price Per Unit"].transform(lambda x: x.fillna(x.mean()))
df["Total Spent"] = df.groupby("Item")["Total Spent"].transform(lambda x: x.fillna(x.mean()))
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"] , errors="coerce")
```

```python
df = df.sort_values("Transaction Date")
df["Transaction Date"]
```

```python
df["Transaction Date"] = df["Transaction Date"].fillna(method="ffill")
```

```python
df["Total Spent"] = df["Quantity"] * df["Price Per Unit"]
```

```python
df.drop("Transaction ID", axis=1 , inplace =True)
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
x = df.drop("Total Spent" , axis = 1)
y = df["Total Spent"]
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2 , random_state = 42)
```

```python
numeric_features = x.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = x.select_dtypes(include=["object", "category"]).columns.tolist()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor

numeric_pipe = Pipeline(steps=[("scaler",StandardScaler())])
categ_pipe  = Pipeline(steps=[("onehot",OneHotEncoder(handle_unknown="ignore"))])

preprocessor = ColumnTransformer(transformers=[("num" , numeric_pipe , numeric_features),("cat" , categ_pipe ,categorical_features )])
```

```python
pipeline = Pipeline(steps=[("preprocessor" , preprocessor) , ("model" ,RandomForestRegressor(random_state=42) )])
```

**GridSearchCV**

```python
param_grid = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [None, 10, 20],
    "model__min_samples_split": [2, 5]
    }

grid = GridSearchCV(pipeline , param_grid , n_jobs=-1 , cv=5 , scoring="r2")
grid.fit(xtrain , ytrain)
```

```python
print("Best Params:", grid.best_params_)
print("Train RÂ²:", grid.best_score_)
print("Test RÂ²:", grid.score(xtest, ytest))
```
